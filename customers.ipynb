Queries

Query 1:How many customers exist per city, and what is their average account balance?


```python
from tabulate import tabulate

query1 = """ SELECT 
    c.city,
    COUNT(DISTINCT c.customer_id) AS total_customers,
    AVG(a.account_balance) AS average_account_balance
FROM customers_data c
JOIN accounts_data a
    ON c.customer_id = a.customer_id
GROUP BY c.city;"""
cursor_pymysql.execute(query1)
result = cursor_pymysql.fetchall()



# Column headers (must match SELECT order)
headers = ["City", "Total Customers", "Average Account Balance"]

# Print as table
print(tabulate(result, headers=headers, tablefmt="grid",floatfmt=".2f"))
```

Query2: Which account type (Savings, Current, Loan, etc.) holds the highest total balance?

```python
from tabulate import tabulate

query = """
SELECT 
    c.account_type,
    SUM(a.account_balance) AS total_balance
FROM customers_data c
JOIN accounts_data a
    ON c.customer_id = a.customer_id
GROUP BY c.account_type
ORDER BY total_balance DESC
LIMIT 1;
"""

cursor_pymysql.execute(query)
result = cursor_pymysql.fetchall()

headers = ["Account Type", "Total Balance"]

print(tabulate(result, headers=headers, tablefmt="grid",floatfmt=".2f"))
```

Query3:Who are the top 10 customers by total account balance across all account types?


```python
from tabulate import tabulate

query = """
SELECT 
    c.customer_id,
    c.name,
    SUM(a.account_balance) AS total_balance
FROM customers_data c
JOIN accounts_data a
    ON c.customer_id = a.customer_id
GROUP BY c.customer_id, c.name
ORDER BY total_balance DESC
LIMIT 10;
"""

cursor_pymysql.execute(query)
result = cursor_pymysql.fetchall()

headers = ["Customer ID", "Customer Name", "Total Balance"]

print(tabulate(result, headers=headers, tablefmt="grid",floatfmt=".2f"))
```

Query4:Which customers opened accounts in 2023 with a balance above ₹1,00,000?


```python
from tabulate import tabulate

query = """
SELECT 
    c.customer_id,
    c.name,
    c.join_date,
    SUM(a.account_balance) AS total_balance
FROM customers_data c
JOIN accounts_data a
    ON c.customer_id = a.customer_id
WHERE YEAR(c.join_date) = 2023
GROUP BY c.customer_id, c.name, c.join_date
HAVING SUM(a.account_balance) > 100000
ORDER BY total_balance DESC;
"""

cursor_pymysql.execute(query)
result = cursor_pymysql.fetchall()
formatted_result = [
    (r[0], r[1], r[2], f"{float(r[3]):.2f}")
    for r in result
]

headers = ["Customer ID", "Customer Name", "Join Date", "Account Balance"]
print(tabulate(formatted_result , headers=headers, tablefmt="grid", floatfmt=".2f"))
```

Query5:What is the total transaction volume (sum of amounts) by transaction type?


```python
from tabulate import tabulate

query = """
SELECT
    txn_type,
    SUM(amount) AS total_transaction_volume
FROM transactions_data
GROUP BY txn_type
ORDER BY total_transaction_volume DESC;
"""

cursor_pymysql.execute(query)
result = cursor_pymysql.fetchall()

headers = ["Transaction Type", "Total Transaction Volume"]

print(tabulate(
    result,
    headers=headers,
    tablefmt="grid",
    floatfmt=".2f"   # prevents scientific notation
))
```

Query6:How many failed transactions occurred for each transaction type?


```python
from tabulate import tabulate

query = """
SELECT
    txn_type,
    COUNT(*) AS failed_transaction_count
FROM transactions_data
WHERE status = 'failed'
GROUP BY txn_type
ORDER BY failed_transaction_count DESC;
"""

cursor_pymysql.execute(query)
result = cursor_pymysql.fetchall()

headers = ["Transaction Type", "Failed Transactions"]

print(tabulate(
    result,
    headers=headers,
    tablefmt="grid"
))
```

 Query8: Which accounts have 5 or more high-value transactions above ₹20,000?

```python
from tabulate import tabulate

query = """
SELECT
    txn_id,
    COUNT(*) AS high_value_transaction_count
FROM transactions_data
WHERE amount > 20000
GROUP BY txn_id
HAVING COUNT(*) >= 5
ORDER BY high_value_transaction_count DESC;
"""

cursor_pymysql.execute(query)
result = cursor_pymysql.fetchall()

headers = ["Account ID", "High Value Transactions"]

print(tabulate(
    result,
    headers=headers,
    tablefmt="grid"
))
```

Query9:What is the average loan amount and interest rate by loan type (Personal, Auto, Home, etc.)?


```python
from tabulate import tabulate

query = """
SELECT
    Loan_Type,
    AVG(Loan_Amount) AS Avg_Loan_Amount,
    AVG(Interest_Rate) AS Avg_Interest_Rate
FROM loan_amount_data
GROUP BY Loan_Type
ORDER BY Avg_Loan_Amount DESC;
"""

cursor_pymysql.execute(query)
result = cursor_pymysql.fetchall()

headers = ["Loan Type", "Avg Loan Amount", "Avg Interest Rate (%)"]

print(tabulate(
    result,
    headers=headers,
    tablefmt="grid",
    floatfmt=".2f"   # ensures 2 decimal places, no scientific notation
))
```

Query10:Which customers currently hold more than one active or approved loan?

```python
from tabulate import tabulate

query = """
SELECT
    customer_id,
    COUNT(loan_id) AS active_approved_loans
FROM loan_amount_data
WHERE LOWER(loan_status) IN ('active', 'approved')
GROUP BY customer_id
HAVING COUNT(loan_id) > 1
ORDER BY active_approved_loans DESC;
"""

cursor_pymysql.execute(query)
result = cursor_pymysql.fetchall()

headers = ["Customer ID", "Active/Approved Loans"]

print(tabulate(result, headers=headers, tablefmt="grid"))
```

Query11:Who are the top 5 customers with the highest outstanding (non-closed) loan amounts?

Query13: How many customers exist in each age group (e.g., 18–25, 26–35, etc.)?

```python
from tabulate import tabulate

query = """
SELECT
    CASE
        WHEN age BETWEEN 18 AND 25 THEN '18–25'
        WHEN age BETWEEN 26 AND 35 THEN '26–35'
        WHEN age BETWEEN 36 AND 45 THEN '36–45'
        WHEN age BETWEEN 46 AND 55 THEN '46–55'
        WHEN age BETWEEN 56 AND 65 THEN '56–65'
        ELSE '66+'
    END AS age_group,
    COUNT(*) AS customer_count
FROM customers_data
GROUP BY age_group
ORDER BY
    CASE age_group
        WHEN '18–25' THEN 1
        WHEN '26–35' THEN 2
        WHEN '36–45' THEN 3
        WHEN '46–55' THEN 4
        WHEN '56–65' THEN 5
        ELSE 6
    END;
"""

cursor_pymysql.execute(query)
result = cursor_pymysql.fetchall()

headers = ["Age Group", "Number of Customers"]

print(tabulate(result, headers=headers, tablefmt="grid"))
```

Query14:Which issue categories have the longest average resolution time?


```python
from tabulate import tabulate

query = """
SELECT 
    Issue_Category,
    AVG(DATEDIFF(Date_Closed, Date_Opened)) AS avg_resolution_days
FROM support_tickets_data
WHERE Date_Closed IS NOT NULL
GROUP BY Issue_Category
ORDER BY avg_resolution_days DESC;
"""

cursor_pymysql.execute(query)
result = cursor_pymysql.fetchall()

headers = ["Issue categories", "Longer resolution time"]

print(tabulate(result, headers=headers, tablefmt="grid"))
```

Query15:Which support agents have resolved the most critical tickets with high customer ratings (≥4)?


```python
from tabulate import tabulate

query = """
SELECT
    Support_Agent,
    COUNT(*) AS critical_high_rating_tickets
FROM support_tickets_data
WHERE Priority = 'Critical'
  AND Status = 'Closed'
  AND Customer_Rating >= 4
GROUP BY Support_Agent
ORDER BY critical_high_rating_tickets DESC;
"""

cursor_pymysql.execute(query)
result = cursor_pymysql.fetchall()

headers = ["Support Agent", "Critical Tickets (Rating ≥ 4)"]

print(tabulate(result, headers=headers, tablefmt="grid"))
```
